# Feature Engineering 

```python
features = df.drop(['Country Name_left','geometry','Country Name_right','Year','Intermittent preventive treatment (IPT) of malaria in pregnancy (% of pregnant women)','Gi_Star_Incidence','Gi_Star_Mortality'],axis=1)
```

```python
features.info()
```

# TRAIN TEST SPLIT

```python
# Set the target for the prediction
target='Incidence of malaria (per 1,000 population at risk)'


# Create arrays for the features and the response variable

# set X and y
y = features[target]
X = features.drop(target, axis=1)
from sklearn.model_selection import train_test_split
# Split the data set into training and testing data sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python

lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)

y_pred
```

```python
lr.score(X_train, y_train)
```

# Evaluation 

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

## Implementing Ridge Regression

```python
from sklearn.linear_model import Ridge, RidgeCV

alphas = np.random.uniform(0, 10, 50)
ridge_cv = RidgeCV(alphas = alphas, cv = 10, normalize = True)
ridge_cv.fit(X_train, y_train)
```

```python
# best alpha parameter

alpha = ridge_cv.alpha_
alpha
```

```python
ridge = Ridge(alpha = ridge_cv.alpha_)
ridge.fit(X_train, y_train)
```

```python
ridge.score(X_train, y_train)
```

```python
y_pred_ridge=ridge.predict(X_test)

y_pred_ridge
```

```python
r2 = r2_score(y_test,y_pred_ridge)
print(r2)
```

```python
mean_absolute_error(y_test,y_pred_ridge)
```

```python
mean_squared_error(y_test,y_pred_ridge)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred_ridge))
```
